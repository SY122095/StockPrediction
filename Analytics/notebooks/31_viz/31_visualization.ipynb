## 0. 環境セットアップ

```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import japanize_matplotlib

sys.path.append(str(Path('../../').resolve()))
from funcs import data_loader, visualization

pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
cfg     = data_loader.load_yaml('../../configs/configs.yaml')
viz_cfg = visualization.setup_matplotlib('../../configs/visualize.yaml', theme='presentation')
# ↑ rcParams をグローバルに適用済み（フォントサイズ・グリッド・線幅など）

PROCESSED_DIR = Path(cfg['paths']['processed_data'])
FIGURES_DIR   = Path(cfg['paths']['figures'])
```

```python
# --- visualize.yaml の設定値を変数として展開 ---
# ここを変えるだけで全グラフに反映される

_mpl_cfg   = viz_cfg.get('visualization', {}).get('matplotlib', {})
_theme_key = viz_cfg.get('visualization', {}).get('theme', 'presentation')
_theme_cfg = viz_cfg.get('theme', {}).get(_theme_key, {})
_sns_cfg   = viz_cfg.get('visualization', {}).get('seaborn', {})
_pal_cfg   = viz_cfg.get('palette', {})

# 図サイズ・解像度
FIGSIZE      = tuple(_mpl_cfg.get('figure', {}).get('figsize', [12, 6]))   # 標準サイズ
FIGSIZE_WIDE = (FIGSIZE[0], FIGSIZE[1])                                     # 横長（標準と同じ）
FIGSIZE_SQ   = (FIGSIZE[1], FIGSIZE[1])                                     # 正方形（相関・散布図用）
DPI          = _mpl_cfg.get('figure', {}).get('dpi', 150)

# カラーパレット
COLORS   = _pal_cfg.get('categorical',
           ['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2',
            '#937860','#DA8BC3','#8C8C8C','#CCB974','#64B5CD'])
CMAP_SEQ = _pal_cfg.get('sequential', 'Blues')      # 連続値ヒートマップ
CMAP_DIV = _pal_cfg.get('diverging',  'RdBu_r')     # ダイバージングヒートマップ
PALETTE  = _sns_cfg.get('palette', 'deep')           # seaborn パレット

# 線・点
LW      = _mpl_cfg.get('lines', {}).get('linewidth', 2)
ALPHA   = _mpl_cfg.get('scatter', {}).get('alpha', 0.7)

print(f'theme     : {_theme_key}')
print(f'FIGSIZE   : {FIGSIZE}')
print(f'DPI       : {DPI}')
print(f'COLORS[0] : {COLORS[0]}')
print(f'CMAP_SEQ  : {CMAP_SEQ}')
```
