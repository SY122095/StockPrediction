## チェックリスト（受領データチェックシート準拠）

| カテゴリ | 項目 | 内容 | 結果 |
|---------|------|------|------|
| 基本情報 | ファイル名 | 事前共有と一致しているか | ☐ |
| 基本情報 | ファイル形式 | CSV/Excel/SQLiteなど想定どおり | ☐ |
| 基本情報 | ファイルサイズ | 異常に小さくないか | ☐ |
| 基本情報 | ファイル数 | 想定の数と一致しているか | ☐ |
| 基本情報 | パスワード | 共有されたパスワードで開けるか | ☐ |
| データ構造 | レコード数 | 想定と一致 or 異常に少なくないか | ☐ |
| データ構造 | カラム数 | 想定された項目数と一致 | ☐ |
| データ構造 | 主キー候補 | 重複がないか | ☐ |
| データ構造 | 時系列 | 必要期間が揃っているか | ☐ |
| 欠損値 | 欠損の有無 | NULL/空欄/NaNの存在確認 | ☐ |
| 欠損値 | 欠損率 | 特定カラムが異常に欠損していないか | ☐ |
| 欠損値 | 必須項目 | ID・日付・金額などが欠損していないか | ☐ |
| データ型 | 数値型 | 文字列扱いの数値がないか | ☐ |
| データ型 | 日付型 | 変換不可の日付がないか | ☐ |
| データ型 | カテゴリ型 | コード体系の混在なし | ☐ |
| データ型 | コード形式 | コード桁数が統一されているか | ☐ |
| 異常値 | 外れ値 | 極端に高い/低い値の確認 | ☐ |
| 異常値 | 不正値 | 物理的にあり得ない値（年齢200歳など） | ☐ |
| 異常値 | 負の値 | 売上・数量などに負の値がないか | ☐ |
| 重複 | 主キー重複 | ID等の重複確認 | ☐ |
| 重複 | 完全一致行 | 行丸ごとの重複有無 | ☐ |
| 論理整合性 | マスタ参照 | 存在しないコードが使われていないか | ☐ |
| 論理整合性 | 日付順序 | 日付順が破綻していないか | ☐ |
| 論理整合性 | 金額整合性 | 単価×数量=金額か | ☐ |
| メタ情報 | データ辞書 | 項目定義が提供されているか | ☐ |
| メタ情報 | 変数説明 | カラムの意味が明確か | ☐ |
| メタ情報 | 更新日 | データ更新日時の確認 | ☐ |
| セキュリティ | 個人情報 | 必要以上の個人情報が含まれていないか | ☐ |
